```python
platforms = bs_esrb.find('div',{'class':'platforms'}).find('div',{'class':'platforms-txt'})
platforms = platforms.text.replace(',',"")
platform_array = platforms.split()
all_platforms = ['ps4','ps5','xbox_one','xbox_series','pc','nintendo_switch']
platforms_in_game = []
for index, platform in enumerate(platform_array):
    if(platform == 'Xbox'):
        if(platform_array[index+1] == 'One'):
            platforms_in_game.append('xbox_one')
        elif(platform_array[index+1] == 'Series'):
            platforms_in_game.append('xbox_series')
    elif(platform == 'PlayStation'):
        if(platform_array[index+1] == '4'):
            platforms_in_game.append('ps4')
        elif(platform_array[index+1] == '5'):
            platforms_in_game.append('ps5')
    elif(platform == 'Nintendo'):
        if(platform_array[index+1] == 'Switch'):
            platforms_in_game.append('nintendo_switch')
    elif(platform == 'PC'):
        platforms_in_game.append('pc')
platforms_not_in_game = [x for x in all_platforms if x not in platforms_in_game]
for platform in platforms_in_game:
    if platform == 'ps4':
        ps4.append(1)
    if platform == 'ps5':
        ps5.append(1)
    if platform == 'xbox_one':
        xbox_one.append(1)
    if platform == 'xbox_series':
        xbox_series.append(1)
    if platform == 'pc':
        pc.append(1)
    if platform == 'nintendo_switch':
        nintendo_switch.append(1)
for platform in platforms_not_in_game:
    if platform == 'ps4':
        ps4.append(0)
    if platform == 'ps5':
        ps5.append(0)
    if platform == 'xbox_one':
        xbox_one.append(0)
    if platform == 'xbox_series':
        xbox_series.append(0)
    if platform == 'pc':
        pc.append(0)
    if platform == 'nintendo_switch':
        nintendo_switch.append(0)
```
